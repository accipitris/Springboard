```python
## 5.2.2 Python Exercises
## Question 1) Find the 10 countries with most projects
## Question 2) Find the top 10 major project themes (using column 'mjtheme_namecode'),
## Question 3) In 2. above you will notice that some entries have only the code and the name is missing,
## Create a dataframe with the missing names filled in\n",

import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
# load json as string
file = open('C:/Users/jlpar/Documents/Springboard Data Files/world_bank_projects.json')
data = json.load(file)

## Normalize the data
data_norm = pd.io.json.json_normalize(data)
data

## Putting JSON File into a dataframe
# load as Pandas dataframe
data_df = pd.read_json('C:/Users/jlpar/Documents/Springboard Data Files/world_bank_projects.json')
data_df.head(3)
```

```python
## Selecting columns of interest with a re-indexing tool for this table
data_df_I = data_df.reindex(columns=['id','country_namecode', 'countrycode', 'countryname', 'countryshortname', 
                                     'project_name','mjtheme_namecode'])
data_df_I.head(3)
```

```python
## 5.2.2 Python Exercises
## Question 1) Find the 10 countries with most projects with the value and sort counts for countryname and 
## grouping by countrycode with the top ten projects listed first. 

data_df_topcountry = data_df.groupby('countrycode')['countryname'].value_counts().sort_values(ascending=False).head(10)
print('Top Ten Countries with the most projects:')
print(data_df_topcountry.head(10))
```

```python
## Questions 2 and 3 
## Normalizing the data as shown in the handout example for the 'mjtheme_namecode'.  This contains the code for the 
## Project and the name of the project.  This printout shows the missing value for the name which gives a differnce
## of counts for the value_counts of 'code' and 'name'.

data_norm2 = json_normalize(data, 'mjtheme_namecode', ['countryname'])
print(data_norm2.head(3))
```

```python
## 5.2.2 Python Exercises
## Question 2) Find the top 10 major project themes (using column 'mjtheme_namecode')

## The value_counts was used on the name of the project with the sorting_values function listing
## the top projects and grouping on 'code'.  

data_norm2.groupby('code').name.value_counts().sort_values(ascending=False).head(10)
```

```python
## Value counts was also applied to the 'code' group using the value_counts and sort_values with the top
## values.  This list shows a difference between the name and code attributes by a value of 27.  This is the reason 
## why question 3 wants missing values replaced for the project name.  

data_norm2.code.value_counts().sort_values(ascending=False).head(10)
```

```python
## Normalize the data as shown in the Python Example, Replacing Empty values with NaN, Sort the values
## based on 'code' and then use the forward fill method in python to replace the empty values '' with NaN. 

data_r = json_normalize(data, 'mjtheme_namecode', ['countryname'])
data_r['name'].replace('', np.nan, inplace=True)
data_rs = data_r.sort_values(by='code')
```

```python
## Question 3) In 2. above you will notice that some entries have only the code and the name is missing,
## Create a dataframe with the missing names filled in\n",

## When the missing values in the 'name' section was replaced with 'NaN', the top ten projects based on
## code and name match at 250 for this assignment.  When compared to the previous outputs, the top 10 code starts
# with 250 while the top project only had 223.

data_rs['name'].fillna(method='ffill', inplace=True)
data_rs.groupby('code').name.value_counts().sort_values(ascending=False).head(10)
```
